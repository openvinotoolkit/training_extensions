```python
from otx.core.data.module import OTXDataModule
from otx.core.engine.engine import Engine
from otx.core.model.module.base import OTXLitModule

config_file = "test_config.yaml"

datamodule = OTXDataModule.from_config(config=config_file)
model = OTXLitModule.from_config(config=config_file)
engine = Engine.from_config(config=config_file)

engine.train(model=model, datamodule=datamodule, max_epochs=3)
```
