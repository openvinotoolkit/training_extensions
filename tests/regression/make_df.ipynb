```python
from tests.regression.regression_test_helpers import(
    TEST_TYPES,
    TASK_TYPES,
    TRAIN_TYPES,
    LABEL_TYPES
)

import json
import pandas as pd
import os


def get_metric_dict(dict_type, idx, model):
    print(dict_type, idx, model)
    if dict_type:
        return dict_type[idx][model]
    else:
        return "-"

def get_metric_items(input_data):
    if isinstance(input_data, dict):
        return_list = []
        for k, v in input_data.items():
            return_list.append((k, v))
        return return_list
    else:
        return "-"
    
root_path = "/tmp/regression_test_results/"

result_data = {
    'Task type': [], 
    'Train type': [], 
    'Label type': [], 
    'Model': [], 
    "Train": [],
    "Export": [],
    "Deploy": [],
    "NNCF": [],
    "POT": [],

    "Train E2E Time (Sec.)": [],
    "Export E2E Time (Sec.)": [],
    "Deploy E2E Time (Sec.)": [],
    "NNCF E2E Time (Sec.)": [],
    "POT E2E Time (Sec.)": [],

    "Train Eval Time (Sec.)": [],
    "Export Eval Time (Sec.)": [],
    "Deploy Eval Time (Sec.)": [],
    "NNCF Eval Time (Sec.)": [],
    "POT Eval Time (Sec.)": []
}
for root, dirs, files in os.walk(root_path):
    for file in files:
        task = root.split('/')[-1]
        json_file_path = os.path.join(root, file)            
        if "tiling" in task:
            task_key = '_'.join(task.split('_')[1:])
        else:
            task_key = task
        
        if "anomaly" not in task:
            with open(json_file_path, "r") as f:
                data = json.load(f)
            
            for label_type in LABEL_TYPES:
                for train_type in TRAIN_TYPES:
                    print(task, label_type, train_type, task_key)
                    train_type_data = data[task_key][label_type][train_type]
                    train_values = train_type_data.get('train', None) 
                    export_values = train_type_data.get('export', None) 
                    deploy_values = train_type_data.get('deploy', None) 
                    nncf_values = train_type_data.get('nncf', None) 
                    pot_values = train_type_data.get('pot', None) 
                    
                    if train_values is not None:
                        print('[*] label_type: ', label_type)
                        print('[*] train_type: ', train_type)
                        for i, train_value in enumerate(train_values):
                            for model, performances in train_value.items():
                                train_items = get_metric_items(get_metric_dict(train_values, i, model))
                                export_items = get_metric_items(get_metric_dict(export_values, i, model))
                                deploy_items = get_metric_items(get_metric_dict(deploy_values, i, model))
                                nncf_items = get_metric_items(get_metric_dict(nncf_values, i, model))
                                pot_items = get_metric_items(get_metric_dict(pot_values, i, model))
                                
                                result_data['Task type'].append(task)
                                result_data['Train type'].append(train_type)
                                result_data['Label type'].append(label_type)
                                result_data['Model'].append(model)
                                
                                if isinstance(train_items, list):
                                    result_data['Train'].append(f"{train_items[0][0]}: {train_items[0][1]}") 
                                    result_data['Train E2E Time (Sec.)'].append(f"{train_items[2][1]}")
                                    result_data['Train Eval Time (Sec.)'].append(f"{train_items[3][1]}")
                                else:
                                    result_data['Train'].append(train_items)
                                    result_data['Train E2E Time (Sec.)'].append(train_items)
                                    result_data['Train Eval Time (Sec.)'].append(train_items)
                                
                                if isinstance(export_items, list):
                                    result_data['Export'].append(f"{export_items[0][0]}: {export_items[0][1]}") 
                                    result_data['Export E2E Time (Sec.)'].append(f"{export_items[1][1]}")
                                    result_data['Export Eval Time (Sec.)'].append(f"{export_items[2][1]}")
                                else:
                                    result_data['Export'].append(export_items) 
                                    result_data['Export E2E Time (Sec.)'].append(export_items)
                                    result_data['Export Eval Time (Sec.)'].append(export_items)
                                
                                if isinstance(deploy_items, list):
                                    result_data['Deploy'].append(f"{deploy_items[0][0]}: {deploy_items[0][1]}")
                                    result_data['Deploy E2E Time (Sec.)'].append(f"{deploy_items[1][1]}")
                                    result_data['Deploy Eval Time (Sec.)'].append(f"{deploy_items[2][1]}")
                                else:
                                    result_data['Deploy'].append(deploy_items)
                                    result_data['Deploy E2E Time (Sec.)'].append(deploy_items)
                                    result_data['Deploy Eval Time (Sec.)'].append(deploy_items)
                                    
                                if isinstance(nncf_items, list):
                                    result_data['NNCF'].append(f"{nncf_items[0][0]}: {nncf_items[0][1]}")
                                    result_data['NNCF E2E Time (Sec.)'].append(f"{nncf_items[1][1]}")
                                    result_data['NNCF Eval Time (Sec.)'].append(f"{nncf_items[2][1]}")
                                else:
                                    result_data['NNCF'].append(nncf_items)
                                    result_data['NNCF E2E Time (Sec.)'].append(nncf_items)
                                    result_data['NNCF Eval Time (Sec.)'].append(nncf_items)
                                    
                                if isinstance(pot_items, list):
                                    result_data['POT'].append(f"{pot_items[0][0]}: {pot_items[0][1]}")
                                    result_data['POT E2E Time (Sec.)'].append(f"{pot_items[1][1]}")
                                    result_data['POT Eval Time (Sec.)'].append(f"{pot_items[2][1]}")
                                else:
                                    result_data['POT'].append(pot_items)
                                    result_data['POT E2E Time (Sec.)'].append(pot_items)
                                    result_data['POT Eval Time (Sec.)'].append(pot_items)

df = pd.DataFrame(result_data)
df.head()
```
